```python
#1.Random Forest
import pandas as pd
import numpy as np

#RandomForestClassifier algorithm
#Assigning Input Column Features(Data) to 'X' and Output Column(Target) to 'Y'..
X=np.array(horizontal_stack.iloc[:,1:260]) #Input Cols
X 
Y=np.array(horizontal_stack.iloc[:,260]) #Output Cols
Y
```

```python
#Random Splitting of data..
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.25) #Train data size 70% and Test data size 30%
X_train,X_test,Y_train,Y_test
```

```python
#Importing some specific packages..
from sklearn.ensemble import RandomForestClassifier

#Random Forest Classifier function..
rfc = RandomForestClassifier(n_estimators=600,bootstrap=True,n_jobs=-1,random_state=42)
rfc
```

```python
#Fitting/Training the model..
rfclftrain = rfc.fit(X_train,Y_train)
rfclftrain
```

```python
#Prediction the model on test data..
rfclftest = rfclftrain.predict(X_test)
rfclftest
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix #Importing on evaluation metrics..

#Evaluating some confusion matrix,score on test data..
confusion_matrix(rfclftest,Y_test)
accuracy_score(rfclftest,Y_test) #100
```

```python
#Prediction the model on train data..
rfclftrain = rfclftrain.predict(X_train)
rfclftrain
```

```python
#Evaluating some confusion matrix,score on train data..
confusion_matrix(rfclftrain,Y_train)
accuracy_score(rfclftrain,Y_train) #100
```

```python
#2.Support Vector Machine (SVM) algorithm
#SVM Package.. 
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split


train,test = train_test_split(horizontal_stack,test_size = 0.25)
train,test
```

```python
train_X = train.iloc[:,1:260]
train_Y = train.iloc[:,260]
test_X = test.iloc[:,1:260]
test_Y = test.iloc[:,260]
```

```python
#Implemeting Kernel Function..
#Simple linear function
simple_linear = SVC(kernel = "linear")
simple_linear.fit(train_X,train_Y)
testingX = simple_linear.predict(test_X)

np.mean(testingX == test_Y) #100
```

```python
#Sigmoid function
simple_sig = SVC(kernel = "sigmoid")
simple_sig.fit(train_X,train_Y)
testingX3 = simple_sig.predict(test_X)

np.mean(testingX3 == test_Y) #66.6667
```

```python
#Poly function
simple_poly = SVC(kernel = "poly")
simple_poly.fit(train_X,train_Y)
testingX2 = simple_poly.predict(test_X)

np.mean(testingX2 == test_Y) #100
```

```python
#RBF function
simple_rbf = SVC(kernel = "rbf")
simple_rbf.fit(train_X,train_Y)
testingX1 = simple_rbf.predict(test_X)

np.mean(testingX1 == test_Y) #100
```
